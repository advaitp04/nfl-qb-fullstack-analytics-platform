```python
#Calculate valuable totals and condense the QB table 
qb_season_stats = (
    qb_stats_23.groupby(
        ["player_id", "player_name", "player_display_name", "headshot_url", "season", "team"],
        as_index=False
    )
    .agg(
        total_passing_yds=("passing_yards", "sum"),
        total_passing_tds=("passing_tds", "sum"),
        total_rushing_tds=("rushing_tds", "sum"),
        total_attempts=("attempts", "sum"),
        total_completions=("completions", "sum"),
        total_carries=("carries", "sum"),
        total_ints=("passing_interceptions", "sum"),
        total_sack_fumbles=("sack_fumbles", "sum"),
        total_sack_fumbles_lost=("sack_fumbles_lost", "sum"),
        total_rushing_fumbles=("rushing_fumbles", "sum"),
        total_rushing_fumbles_lost=("rushing_fumbles_lost", "sum"),
        total_sacks=("sacks_suffered", "sum"),
        total_sack_yds_lost=("sack_yards_lost", "sum"),
        total_air_yds=("passing_air_yards", "sum"),
        total_epa=("passing_epa", "sum"),
        total_first_downs=("passing_first_downs", "sum"),
        total_2pt_passing_convs=("passing_2pt_conversions", "sum"),
        total_2pt_rushing_convs=("rushing_2pt_conversions", "sum"),
        total_fantasy_points=("fantasy_points", "sum"),
    )
)
```

```python
#dropback column for simplifying further calculations 
qb_season_stats["total_dropbacks"] = qb_season_stats['total_attempts'] + qb_season_stats['total_sacks']
qb_season_stats["total_fumbles_lost"] = qb_season_stats['total_sack_fumbles_lost'] + qb_season_stats['total_rushing_fumbles_lost']
qb_season_stats["total_tds"] = qb_season_stats['total_passing_tds'] + qb_season_stats['total_rushing_tds']

#Turnover Risk factors 
qb_season_stats["int_rate"] = qb_season_stats['total_ints'] / qb_season_stats['total_attempts']
qb_season_stats["fumble_lost_rate"] =  qb_season_stats['total_fumbles_lost'] / qb_season_stats['total_dropbacks']

#Drive Sustainability
qb_season_stats["first_down_rate"] = qb_season_stats['total_first_downs'] / qb_season_stats['total_attempts']
qb_season_stats["completion_rate"] = qb_season_stats['total_completions'] / qb_season_stats['total_attempts']
qb_season_stats["sack_rate"] = qb_season_stats['total_sacks'] / qb_season_stats['total_dropbacks']
qb_season_stats["sack_yards_per_sack"] = qb_season_stats['total_sack_yds_lost'] / qb_season_stats['total_sacks'].replace(0, np.nan)

#Offensive Success
qb_season_stats["td_per_attempt"] = qb_season_stats['total_tds'] / qb_season_stats['total_attempts']
qb_season_stats['epa_per_dropback'] = qb_season_stats['total_epa'] / qb_season_stats['total_dropbacks']
qb_season_stats['yards_per_attempt'] = qb_season_stats['total_passing_yds'] / qb_season_stats['total_attempts']
```

```python
#Normalize all metrics and invert bad ones to ensure each metric value falls between 0 and 1, making each factor's contribution to the cortisol score proportional
good_metrics = [
    "first_down_rate",
    "completion_rate",
    "td_per_attempt",
    "epa_per_dropback",
    "yards_per_attempt"
]

bad_metrics = [
    "int_rate",
    "fumble_lost_rate",
    "sack_rate",
    "sack_yards_per_sack"
]

for good_metric in good_metrics:
    qb_season_stats[good_metric + "_norm"] = (qb_season_stats[good_metric] - min(qb_season_stats[good_metric])) / (max(qb_season_stats[good_metric]) - min(qb_season_stats[good_metric]))

for bad_metric in bad_metrics:
    qb_season_stats[bad_metric + "_norm"] = 1 - ((qb_season_stats[bad_metric] - min(qb_season_stats[bad_metric])) / (max(qb_season_stats[bad_metric]) - min(qb_season_stats[bad_metric])))
```

```python
#Calculation of cortisol scores for each qb 
rows = []
turnover_factors = ["int_rate_norm", "fumble_lost_rate_norm", "sack_rate_norm", "sack_yards_per_sack_norm"]
drive_factors = ["first_down_rate_norm", "completion_rate_norm"]
success_factors = ["td_per_attempt_norm", "epa_per_dropback_norm", "yards_per_attempt_norm"]

#Compute score values & averaging all three metrics to ensure turnovers don't dominate the final cortisol score
qb_normalized_season_stats["turnover_score"] = qb_normalized_season_stats[turnover_factors].mean(axis=1)
qb_normalized_season_stats["drive_score"] = qb_normalized_season_stats[drive_factors].mean(axis=1)
qb_normalized_season_stats["success_score"] = qb_normalized_season_stats[success_factors].mean(axis=1)

#Compute final cortisol values
qb_normalized_season_stats["cortisol_score"] = (
    0.4 * qb_normalized_season_stats["turnover_score"] +
    0.3 * qb_normalized_season_stats["drive_score"] +
    0.3 * qb_normalized_season_stats["success_score"]
)

#Create rank column based on final cortisol score from highest (most stable) to least stable qbs
qb_normalized_season_stats["rank"] = qb_normalized_season_stats["cortisol_score"].rank(
    ascending=False,
    method="min"
)

#create new dataframe from the newly created values
qb_cortisol_scores = qb_normalized_season_stats[
    [
        "player_id",
        "player_display_name",
        "team",
        "season",
        "turnover_score",
        "drive_score",
        "success_score",
        "cortisol_score",
        "rank"
    ]
].copy()

#sorting
qb_cortisol_scores = qb_cortisol_scores.sort_values(
    by="cortisol_score",
    ascending=False
)


#Change scoring range from 0-100 for better understanding
qb_cortisol_scores['cortisol_score'] = 100 * (qb_cortisol_scores['cortisol_score'] - qb_cortisol_scores['cortisol_score'].min()) / (qb_cortisol_scores['cortisol_score'].max() - qb_cortisol_scores['cortisol_score'].min())
qb_cortisol_scores['cortisol_score'] = qb_cortisol_scores['cortisol_score'].round(1)

qb_cortisol_scores = qb_cortisol_scores.rename(columns={'player_display_name':'qb_name'})

qb_cortisol_scores
```

```python
#Jake Browning, a backup QB that played 10 games in the 2023 season and recorded 267 dropbacks, relatively low compared to the other QBs yet he still ranked #3 in terms of cortisol score
#adjustment below aims to move outliers like Browning closer to the league average 

avg_cortisol_score = qb_cortisol_scores['cortisol_score'].mean()
K = 400

qb_cortisol_scores['total_dropbacks'] = qb_season_stats['total_dropbacks']
qb_cortisol_scores['stabilized_cortisol_score'] = (
    (qb_cortisol_scores["total_dropbacks"] /
    (qb_cortisol_scores["total_dropbacks"] + K)) *
    qb_cortisol_scores["cortisol_score"]
) + (
    (K /
    (qb_cortisol_scores["total_dropbacks"] + K)) *
    avg_cortisol_score
)

qb_cortisol_scores = qb_cortisol_scores.sort_values(
    by="stabilized_cortisol_score",
    ascending=False
)

qb_cortisol_scores["stabilized_rank"] = range(1, len(qb_cortisol_scores)+1)
```

```python
#Filter df by QB's only based on dropbacks (qb_dropback of 1 indicates either a pass, fumble, or sack)
pbp_2023 = pbp_2023[(pbp_2023['season_type']=="REG") & (pbp_2023['qb_dropback']==1)]
print(pbp_2023['passer_player_id'].isna().sum())
print(pbp_2023['id'].isna().sum())
print(pbp_2023['passer_id'].isna().sum())
#Clearly, both id and passer_id are acceptable; however, passer_id is problematic for sacks and fumbles because those two don't count as passes
#therefore, id is a better option
```

```python
#Select subset of columns required for computing each advanced metric
pbp_2023 = pbp_2023.rename(columns={'id':'player_id'})
pbp_2023 = pbp_2023[['player_id', 'season', 'qb_epa', 'play_type','down', 'qb_dropback', 'third_down_converted', 'ydstogo', 'yardline_100', 'pass_touchdown']]
```

```python
pbp_2023
```

```python
#Helper columns for advanced metrics
pbp_2023['is_third_down'] = pbp_2023['down']==3
pbp_2023['is_third_and_long'] =  ((pbp_2023['down'] == 3) & (pbp_2023['ydstogo'] >= 7))
pbp_2023['is_third_and_long_conversion'] = (pbp_2023['is_third_and_long'] & (pbp_2023['third_down_converted']==1))
pbp_2023['is_third_down_conversion'] = (pbp_2023['is_third_down'] & (pbp_2023['third_down_converted']==1))
pbp_2023['is_negative_epa'] = pbp_2023['qb_epa'] < 0
pbp_2023['is_panic_play'] = ((pbp_2023['qb_epa'] < -1) & (pbp_2023['ydstogo'] >= 7) & (pbp_2023['down'].isin([3,4])))
pbp_2023['is_redzone'] = pbp_2023['yardline_100'] <= 20
pbp_2023['is_redzone_td'] = (pbp_2023['is_redzone'] & (pbp_2023['pass_touchdown']==1))
```

```python
advanced_metrics = (
    pbp_2023.groupby(
        ['player_id', 'season'],
        as_index=False
    )
    .agg(
        total_negative_epa_plays = ("is_negative_epa", "sum"),
        total_panic_plays = ("is_panic_play", "sum"),
        total_third_downs = ("is_third_down", "sum"),
        total_third_downs_converted = ("is_third_down_conversion", "sum"),
        total_dropbacks = ("qb_dropback", "count"),
        total_third_and_long = ("is_third_and_long", "sum"),
        total_third_and_long_converted = ("is_third_and_long_conversion", "sum"),
        total_redzone_attempts = ("is_redzone", "sum"),
        total_redzone_tds = ("is_redzone_td", "sum")
    )
)
```

```python
advanced_metrics 
```

```python
#Filter by dropbacks to remove any outliers 
advanced_metrics = advanced_metrics[advanced_metrics['total_dropbacks'] >= 150]
```

```python
#Compute advanced stats
advanced_metrics['negative_epa_rate'] = advanced_metrics['total_negative_epa_plays'] / advanced_metrics['total_dropbacks']
advanced_metrics['panic_play_rate'] = advanced_metrics['total_panic_plays'] / advanced_metrics['total_dropbacks']
advanced_metrics['third_down_conversion_rate'] = advanced_metrics['total_third_downs_converted'] / advanced_metrics['total_third_downs']
advanced_metrics['third_and_long_conversion_rate'] = advanced_metrics['total_third_and_long_converted'] / advanced_metrics['total_third_and_long']
advanced_metrics['third_down_regular_attempts'] = advanced_metrics["total_third_downs"] - advanced_metrics["total_third_and_long"]
advanced_metrics['third_down_regular_conversions'] = advanced_metrics["total_third_downs_converted"] - advanced_metrics["total_third_and_long_converted"]
advanced_metrics['third_down_regular_conversion_rate'] = advanced_metrics["third_down_regular_conversions"] / advanced_metrics["third_down_regular_attempts"]
advanced_metrics['redzone_td_rate'] = advanced_metrics['total_redzone_tds'] / advanced_metrics['total_redzone_attempts']
```
